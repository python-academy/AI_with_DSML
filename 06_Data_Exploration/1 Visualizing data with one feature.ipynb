## Try 6.1.1: Visualizing categorical features in Python.

```python
# Import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
country = pd.read_csv('country_complete.csv')
country
```

```python
# Create a barplot for the number of countries in each continent using histplot
sns.histplot(x='Continent', data=country, shrink=0.8)
```

```python
# Create a barplot for the number of countries in each continent using countplot
sns.countplot(x='Continent', data=country)
plt.ylabel("Count")
```

## Numerical features

```python
# Create a histogram for years of education
sns.histplot(x='Years', data=country)
```

```python
# Create a box plot for years of education
sns.boxplot(x='Years', data=country)
```

```python
# Create a kde plot for years of education
sns.kdeplot(x='Years', data=country)
```

```python
# Create a histogram for years of education with density and kde
sns.histplot(x='Years', data=country, stat='density', kde=True)
```

```python
# Create a violin plot for years of education
sns.violinplot(x='Years', data=country)
```
